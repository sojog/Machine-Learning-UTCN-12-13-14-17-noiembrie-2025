```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# PassengerId: Unique identifier for each passenger.
# Survived: Survival status of the passenger (0 = Not Survived, 1 = Survived).
# Pclass: Passenger class (1 = First class, 2 = Second class, 3 = Third class).
# Sex: Gender of the passenger.
# Age: Age of the passenger.
# SibSp: Number of siblings/spouses aboard the Titanic.
# Parch: Number of parents/children aboard the Titanic.
# Fare: Fare paid by the passenger.
# Embarked: Port of embarkation (C = Cherbourg, Q = Queenstown, S = Southampton).
```

```python
x_y_train = pd.read_csv("titanic/train.csv", index_col="PassengerId")
x_y_train.head(2)
```

```python
# x_y_train["Name"].unique()
```

```python
def drop_columns(df, *columns):
    for col in columns:
        df = df.drop(col, axis=1)
    
    return df
```

```python
x_y_train["FamilySize"] = x_y_train["SibSp"] + x_y_train["Parch"] + 1
x_y_train
```

```python
x_y_train["IsAlone"] = (x_y_train["FamilySize"] == 1).astype(int)
x_y_train
```

```python
x_y_train["Cabin"].isnull().sum()
```

```python
x_y_train["HasCabin"] = x_y_train["Cabin"].notnull()
x_y_train
```

```python
x_y_train["HasCabin"] = x_y_train["Cabin"].notnull().astype(int)
x_y_train
```

```python
x_y_train["Embarked"].isnull().sum()
```

# o tactica este de a umple cu valoarea care apare cel mai des / sau media

```python
x_y_train["Embarked"] = x_y_train["Embarked"].fillna('S')
x_y_train["Embarked"].isnull().sum()
```

```python
x_y_train.isnull().sum()
```

```python
x_y_train["Age"].median()
```

```python
x_y_train["Age"] = x_y_train["Age"].fillna(x_y_train["Age"].median())
x_y_train.isnull().sum()
```

```python
x_y_train.head(2)
```

```python
x_y_train = pd.get_dummies(x_y_train, columns=["Sex", "Embarked"])
x_y_train
```

```python
x_y_train= x_y_train.drop(["Name", "Ticket", "Cabin", "SibSp", "Parch"],  axis=1)
x_y_train
```

```python
x_y_train[['Sex_female', 'Sex_male', 'Embarked_C', 'Embarked_Q',
       'Embarked_S']] = x_y_train[['Sex_female', 'Sex_male', 'Embarked_C', 'Embarked_Q',
       'Embarked_S']].astype(int)
x_y_train
```
