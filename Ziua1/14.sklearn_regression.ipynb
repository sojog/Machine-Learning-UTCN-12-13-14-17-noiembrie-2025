```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import numpy as np
```

```python
model = LinearRegression()
model 
```

```python
## Suprafata si Pret 
apartamente = np.array([
    (35, 50_000),
    (45, 60_000),
    (55, 70_000),
    (60, 75_000),
    (70, 100000),
    (80, 100_000),
    (86, 170_000),
    (90, 200_000),
    (95, 230_000),

])
suprafetele = apartamente[ :, 0]
preturile = apartamente[ :, 1]
```

```python
suprafetele
```

```python
suprafetele_2d = suprafetele.reshape(-1, 1)
suprafetele_2d
```

```python
model.fit(suprafetele_2d, preturile)
```

```python
model.intercept_, model.coef_
```

```python
np.float64(-77623.23943661977), np.float64(2846.7681086519124), 
```

```python

plt.scatter(suprafetele, preturile)
plt.scatter(suprafetele, suprafetele * model.coef_ +  model.intercept_)
plt.plot(suprafetele, suprafetele * model.coef_ +  model.intercept_)
```

```python
model.predict([  [65]  ])
```

```python

plt.scatter(suprafetele, preturile)
plt.scatter(suprafetele, suprafetele * model.coef_ +  model.intercept_)
plt.plot(suprafetele, suprafetele * model.coef_ +  model.intercept_)

mp = 65
plt.scatter(mp, model.predict([  [mp]  ]))


plt.plot(np.ones(100) * mp, np.linspace(20000,  model.predict([  [mp]  ]), 100), 'm--' )
plt.plot(np.linspace(25, mp, 100),  np.ones(100) *  model.predict([  [mp]  ]), 'm--' )
```
