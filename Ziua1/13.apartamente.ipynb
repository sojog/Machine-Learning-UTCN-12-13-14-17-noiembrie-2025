```python
from scipy import stats
```

```python
response = stats.linregress(suprafetele, preturile)
response 
```

```python
response.slope, response.intercept
```

```python
def f(x):
    return response.slope * x +  response.intercept
```

```python
plt.scatter(suprafetele, preturile);

plt.plot(suprafetele, response.slope * suprafetele +  response.intercept, c="black")

plt.plot(suprafetele, suprafetele * 2000, c="red");
plt.plot(suprafetele, suprafetele * 2300, c="green");
plt.plot(suprafetele, suprafetele * 1700, c="blue");
```
