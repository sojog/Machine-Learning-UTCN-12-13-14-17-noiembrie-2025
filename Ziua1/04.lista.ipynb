# Range este tot o colectie

```python
for i in range(10):
    print("i=", i, type(i))
```

```python
print(range(10))
```

```python
for i in range(0, 3):
    print("i=", i, type(i))
```

```python
for i in range(3, 8):
    print("i=", i, type(i))
```

```python
range(10, 20)
```

```python
list(range(10, 20))
```
