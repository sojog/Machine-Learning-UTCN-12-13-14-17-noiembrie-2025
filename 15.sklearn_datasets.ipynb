```python
x, y =  make_regression(n_samples=10, n_features=1)
plt.scatter(x, y, c="orange")
plt.plot(x, y)
```

```python
x, y =  make_regression(n_samples=10, n_features=1, noise=5)
plt.scatter(x, y, c="orange")
plt.plot(x, y)
```

```python
x, y =  make_regression(n_samples=10, n_features=1, noise=5)
plt.scatter(x, y, c="orange")
# plt.plot(x, y)
```

```python
x, y =  make_regression(n_samples=10, n_features=1, noise=5)
model = LinearRegression()
model.fit(x, y)


plt.scatter(x, y, c="orange")
plt.plot(x, model.coef_ * x + model.intercept_ )
```
