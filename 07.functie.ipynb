```python
def adunare(a, b):
    return a + b

def scadere (a, b):
    return a - b

def inmultire(a, b):
    return a * b
```

```python
def operatie(func, a, b):
    return func(a, b)
```

```python
operatie(adunare, 23, 10)
```

```python
operatie(scadere, 23, 10)
```

```python
operatie(inmultire, 23, 10)
```

```python
adunare(30, 20)
```

```python
adunare(30, 20, 320)
```

# O functie poate lua mai multi parametri

```python
def cel_mai_mare_numar(*args):
    print(args, type(args))
    if not args:
        return 0
    return max(args)
```

```python
cel_mai_mare_numar()
```

```python
cel_mai_mare_numar(32)
```

```python
cel_mai_mare_numar(31, 31)
```

```python
cel_mai_mare_numar(31, 31, 3124)
```

```python
def cel_mai_mare_numar_si_cel_mai_mic(*args, **kargs):
    print(args, type(args))
    print(kargs, type(kargs))
```

```python
cel_mai_mare_numar_si_cel_mai_mic()
```

```python
cel_mai_mare_numar_si_cel_mai_mic(321, 312, 321, 312)
```

```python
cel_mai_mare_numar_si_cel_mai_mic(321, 312, 321, 312, a=312, b=31)
```
