```python
recipes = json_obj["recipes"] 
recipes
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame(recipes)
df.head(2) 
```

```python
df = df.drop("id", axis=1)
df.head(2)
```

```python
df.describe()
```

```python
df["difficulty"].unique()
```

```python
df["difficulty"] = df["difficulty"].map({"Easy":0, "Medium":1, "Hard":2})
df.head(2)
```

```python
df["name"].unique(), df["name"].nunique()
```

```python
df["name"].size
```

```python
df["cuisine"].unique(), df["cuisine"].nunique()  
```

```python
df["tags"], type(df["tags"])
```

```python
df["mealType"], type(df["mealType"])
```

```python
df.head(2)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
df["cuisine"] = LabelEncoder().fit_transform(df["cuisine"])
df.head(2)
```

```python
df = df.drop(["instructions", "image","ingredients" ], axis=1)
df.head(2)
```

```python
from sklearn.preprocessing import MultiLabelBinarizer
```

```python
mealTypes = df["mealType"].explode().unique()
mealTypes.sort() 
mealTypes
```

```python
mealType_array =  MultiLabelBinarizer().fit_transform(df["mealType"])
```

```python
df_meals = pd.DataFrame(mealType_array, columns=mealTypes)
df_meals.head(2)
```

```python
tagsType = df["tags"].explode().unique()
tagsType.sort() 
tagsType
```

```python
tagsType.size
```

```python
df.drop(["mealType", "tags" ], axis=1, inplace=True)
```

```python
merged_df = pd.concat([df, df_meals], axis=1)
merged_df.head(2)
```

```python
merged_df.describe()
```
