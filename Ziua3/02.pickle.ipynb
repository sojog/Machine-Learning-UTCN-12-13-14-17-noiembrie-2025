```python
import pickle
```

```python
class Student:
    def __init__(self, nume, varsta):
        self.nume = nume
        self.varsta = varsta

    def __str__(self):
        return f"Studentul are numele {self.nume} si varsta {self.varsta}"
```

```python
obiect_student =  Student("Ionut", 20)
print(obiect_student)
```

```python
obiect_student
```

```python
# scriere in fisier text
with open("fisier_text.txt", "w") as file_handler:
    file_handler.write("Text in fisier")
```

```python
# scriere in fisier binar    -> wb -> write binary
with open("fisier_binar.pkl", "wb") as file_handler:
    pickle.dump(obiect_student, file_handler)
```

```python
# citire din fisier text
with open("fisier_text.txt", "r") as file_handler:
    text = file_handler.read()
    print(text)
```

```python
# citire din fisier binar   -> rb -> read binary
with open("fisier_binar.pkl", "rb") as file_handler:
    obiect_citit = pickle.load(file_handler)
```
