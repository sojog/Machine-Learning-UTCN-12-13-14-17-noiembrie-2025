```python
import pickle
from PIL import Image
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
```

```python
# citire din fisier binar   -> rb -> read binary
with open("model.pkl", "rb") as file_handler:
    ## Python este loosly typed (o variabila poate fi de orice fel de tip de date)
    ## TYPE hint
    obiect_citit:KNeighborsClassifier = pickle.load(file_handler)
```

```python
def read_image(image_path):
    img = Image.open(image_path).convert("L")
    img = np.array(img) / 16
    return img.flatten() 
```

```python
to_predict = read_image("cifra_7.png") 
```

```python
obiect_citit.predict([to_predict])
```
