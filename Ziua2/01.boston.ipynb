```python
scores = {}
for vecin in range(1, 11):
    knn_model = KNeighborsRegressor(n_neighbors=vecin)
    knn_model.fit(x_train, y_train)
    nth_score = knn_model.score(x_test, y_test)
    scores[vecin] = nth_score

print(scores)
```

```python
plt.bar(scores.keys(), scores.values())
```
