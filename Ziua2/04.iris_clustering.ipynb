```python
from sklearn.datasets import load_iris
import numpy as np
import matplotlib.pyplot  as plt
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
iris = load_iris()
type(iris)
```

```python
iris.keys()
```

```python
print(iris["DESCR"])
```

```python
iris["feature_names"]
```

```python
iris["target_names"]
```

```python
x = iris["data"]
y = iris["target"]
```

```python
x.shape
```

```python
sepal_length = x[:, 0]
sepal_width  = x[:, 1]
```

```python
plt.scatter(sepal_length, sepal_width)
plt.xlabel("Sepal length")
plt.ylabel("Sepal width")
```

```python
y
```

```python
plt.scatter(sepal_length, sepal_width, c=y)
plt.xlabel("Sepal length")
plt.ylabel("Sepal width")
plt.legend();
```

```python
petal_length = x[:, 2]
petal_width  = x[:, 3]
```

```python
plt.scatter(petal_length, petal_width, c=y)
plt.xlabel("Petal length")
plt.ylabel("Petal width")
plt.legend();
```

```python
x_train, x_test, y_train, y_test =  train_test_split(x, y, test_size=0.2) 
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn_model =  KNeighborsClassifier()
knn_model
```

```python
knn_model.fit(x_train, y_train)
```

```python
knn_model.score(x_test, y_test)
```

```python
## sepale - train vs test
plt.scatter(x_train[:, 0], x_train[:, 1], c="red")
plt.scatter(x_test[:, 0], x_test[:, 1], c="green")
```

```python
## petale - train vs test
plt.scatter(x_train[:, 2], x_train[:, 3], c="red")
plt.scatter(x_test[:, 2], x_test[:, 3], c="green")
```
